Jauge de contrainte
===================

**Détails sur la Jauge de déformation (Jauge de contrainte)**

https://fr.wikipedia.org/wiki/Jauge_de_d%C3%A9formation

Un contrainte est définie comme suit :
$$ \sigma = F / S $$
F : force de traction ou compression

S : Surface

Loi de Hooke
------------

Un pièce métallique subit une déformation lorsqu'elle est soumise à une force (contrainte)

$$ 
\sigma = E \cdot \epsilon 
$$
$ \sigma $ : contrainte normale

$ \epsilon = {\Delta l \over l} $ : déformation longitudinale

$ E $ : module de Young

Piézorésistance
---------------

Pionnier : Lord Kelvin

$$ R = \rho {l \over A} = \rho {{4 \cdot l} \over {D^2 \cdot \pi}} $$

$\rho$ : résistivité,

$A$ : aire (section) du conducteur,

$l$ : longueur,

$D$ : Diamètre

Si on applique une déformation $\epsilon$ à la longueur, on obtient une variation de la résistance

$$ R + \Delta R = \rho {{l \cdot \epsilon} \over A }$$

```python
R0, epsilon, E, alpha, T0, DL, L0, T=symbols('R0 epsilon E alpha T0 DL L0 T')
DRe=E*R0*epsilon
DRt=alpha*(T-T0)*R0
Rm=R0+DRe+DRt
```

```python
R=R0+DRt
R
```

```python
U0=symbols('U0')
U=U0*(R/(R+Rm)-1/2)
U
```

```python
Us=simplify(U)
```
