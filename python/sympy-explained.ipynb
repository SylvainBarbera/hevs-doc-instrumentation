#  sympy pour les nuls

référence : https://docs.sympy.org/latest/index.html

```python
import sympy as sp
sp.init_printing()
```

Déclarer des variables symboliques

```python
a,b,c,x,y = sp.symbols('a b c x y')
```

Résoudre une équation du type
$$ a+b*x+c*x**2==0 $$

```python
sp.solve(a+b*x+c*x**2,x)
```

Utiliser un résultat

```python
sol=sp.solve(a+b*x+c*x**2,x)
sol[0]
```

Définir une fonction de un ou plusieurs variables

```python
f=sp.Lambda(x, 1+x+x**2)
```

```python
f
```

```python
f(2)
```

Plotter une fonction
- Plot de la fonction pour $$ x \quad in [0,10] $$

```python
sp.plot(f(x),(x,0,10))
```

Calculer une dérivée partielle

```python
f(x).diff(x)
```

Calculer la primitive

```python
f(x).integrate(x)
```

Remplacer une variable

- méthode  subs(var, valeur)

```python
z=a*x+y
```

```python
z.subs(a,2)
```

```python
(a*x+y).subs(a,3)
```

Déclarer une variable avec des propriétés

- On peut spécifier pour un paramètre qu'il est positif. Ceci permet de simplifier les équations.

```python
a1 = sp.symbols('a1')
a2 = sp.symbols('a2', positive=True, real=True)
```

```python
sp.sqrt(a1**2)
```

```python
sp.sqrt(a2**2)
```
